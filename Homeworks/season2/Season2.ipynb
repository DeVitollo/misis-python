```python
#Task 1
class Node (object):
    def __init__ (self,value=None, next_node=None):
        self.value=value
        self.next=next_node
        self.ref=None
        
    def __str__ (self):
        return f"[Node with value {self.value}]"
    
def print_linked_list(head):
    cur = head
    while cur is not None:
        print(cur)
        cur = cur.next
    
def reverse_linked_list (head):
    cur=head
    prev=None
    next=None
    while cur:
        next = cur.next
        cur.next=prev
        prev = cur
        cur=next
    return prev

h, a, b, c, d = Node(1), Node(2), Node(3), Node("Внезапно"), Node(5)

h.next = a
a.next = b
b.next = c
c.next = d

print_linked_list(h)
print('\n')
print('___________________________')
print('       Случилось чудо      ')
print('\n')
print_linked_list(reverse_linked_list(h))
```
