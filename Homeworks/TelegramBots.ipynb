```python
#Bot 1
import telebot
import urllib.request
bot = telebot.TeleBot('token')
@bot.message_handler(content_types=['text'])
def start_message(message):
    if message.text == '/start':
        bot.send_message(message.chat.id, 'Hi! Напиши мне адрес сайта, который  хочешь проверить')
        bot.register_next_step_handler(message, get_url)
    else:
        bot.send_message(message.chat.id, 'Напиши /start')

def get_url(message):
    try:
        webpage = urllib.request.urlopen(message.text)
        if webpage.getcode() == 200:
            bot.send_message(message.chat.id, 'Данный сайт в настоящее время является доступным')
        else:
            bot.send_message(message.chat.id, 'Данный сайт в настоящее время является недоступным')
    except:
        bot.send_message(message.chat.id, 'Неправильно набран адрес')

bot.polling(none_stop=True, interval=0)
```

```python
#Bot 2
import telebot
import re
bot = telebot.TeleBot('token')
@bot.message_handler(content_types=['text'])
def start_message(message):
    if message.text == '/start':
        bot.send_message(message.chat.id, 'Введите анализуруем текст')
        bot.register_next_step_handler(message, analytics)
    else:
        bot.send_message(message.chat.id, 'Напиши /start')
        
def analytics(message):
    global txt
    txt = message.text
    purified  = re.split('@`~#№$%^&[()-;,!?.—\n\t_ ]', txt.lower())
    spisok_vse = [i for i in purified  if not i == '']
    list_ex = ['за', 'про', 'ли', 'так', 'в', 'к', 'а', 'о', 'по', 'под', 'подо', 'ведь', 'ото', 'для','на', 'и', 'но', 'да', 'если', 'как', 'тоже', 'хотя', 'чтобы', 'с', 'то']
    global spisok
    spisok = [word for word in spisok_vse if word not in list_ex]
    bot.send_message(message.chat.id, text = 'Предложений в тексте: ' + str(sentenses()))
    bot.send_message(message.chat.id, text = hype_word())
    bot.send_message(message.chat.id, text = singular_words())
    bot.send_message(message.chat.id, text = longest_word())  
    
def sentenses():
    sentence = txt.count('.') + txt.count('!') + txt.count('?')
    if sentence == 0:
        sentence = 1
    return sentence

def hype_word():
    try:
        count = 0
        word = spisok[0]
        for i in spisok:
            if spisok.count(i) > count:
                count = spisok.count(i)
                word = i
        if count > 1:
            msg1 = f'Самое популярное слово в тексте: {word} \nОно встречается {count} раз(а)'
        else:
            msg1 = 'Уникальными в тексте являются все слова'
    except:
        msg1 = 'В тексте отсутсвуют слова :('  
    return msg1

def singular_words():
    sin = 0
    sin_word = ''
    for i in spisok:
        if spisok.count(i) == 1:
            sin_word = sin_word + i + ", "
            sin = sin + 1
    sin_stat = f'Уникальные слова в тексте: {sin_word} \nЭтих слов аж {sin}'
    return sin_stat

def longest_word():
    long_len = 0
    longest = ''
    for i in spisok:
        if len(i) > long_len:
            long_len = len(i)
            longest = i
    msg2 = f'Самое длинное слово: {longest}\nЕго длина в символах: {long_len}'
    return msg2

bot.polling(none_stop=True, interval=0)
```

```python
#Bot 3
import telebot
from telebot import types
bot = telebot.TeleBot('token')
@bot.message_handler(content_types=['text'])
def start_message(message):
    if message.text == '/start':
        bot.send_message(message.chat.id, 'Введи 2 числа через пробел')
        bot.register_next_step_handler(message, get_calc)
    else:
        bot.send_message(message.chat.id, 'Напиши /start')

def get_calc(message):
    global string_numbers
    string_numbers = message.text.lower()
    calc = types.ReplyKeyboardMarkup(row_width=1)
    item_btn1 = types.KeyboardButton('+')
    item_btn2 = types.KeyboardButton('-')
    item_btn3 = types.KeyboardButton('*')
    item_btn4 = types.KeyboardButton('/')
    item_btn5 = types.KeyboardButton('^')
    calc.add(item_btn1, item_btn2, item_btn3, item_btn4,item_btn5)
    bot.send_message(message.chat.id, 'Определись с математическим оператором:', reply_markup=calc)
    bot.register_next_step_handler(message, calc_do)

def calc_do(message):
    try: 
        counts = string_numbers.split()
        if message.text == '+':
            summ = int(counts[0]) + int(counts[1])
            bot.send_message(message.from_user.id, f'Сумма введеных чисел равна {summ}')
        elif message.text == '-':
            minus = int(counts[0]) - int(counts[1])
            bot.send_message(message.from_user.id, f'Разность введеных чисел равна {minus}')
        elif message.text == '*':
            proizv = int(counts[0]) * int(counts[1])
            bot.send_message(message.from_user.id, f'Произведение введеных чисел равно {proizv}')
        elif message.text == '/':
            if int(counts[1]) == 0:
                bot.send_message(message.from_user.id, 'Деление на 0 невозможно')
            else:
                chast = int(counts[0]) / int(counts[1])
                bot.send_message(message.from_user.id, f'Частное введеных чисел равно {chast}')
        elif message.text == '^':
            stepen = int (counts[0]) ** int(counts[1])
            bot.send_message(message.from_user.id, f'Результат возведения в степень введенных числе равен {stepen}')
    except:
        bot.send_message(message.from_user.id, 'Допушена ошибка. Попробуй заново /start')

    calc = types.ReplyKeyboardRemove(selective=True)
    bot.send_message(message.chat.id, 'Требуется ввести два числа через пробел: ', reply_markup=calc)
    bot.register_next_step_handler(message, get_calc)

    
bot.polling(none_stop=True, interval=0)
```
